```python
import pandas as pd
import numpy as np
import pprint
import datetime

import pickle
```

```python
sub_lgb = pd.read_csv('../data/output/sub_2020-11-14-09-20-22_lightgbm.csv', header=None)
sub_svr = pd.read_csv('../data/output/sub_2020-11-14-10-53-25_svr.csv', header=None)

display(sub_lgb.iloc[:10])
display(sub_svr.iloc[:10])
```

```python
submission = pd.merge(sub_lgb, sub_svr, on=0).drop(0, axis=1).mean(axis='columns')
display(submission)
```

```python
submission = pd.concat([sub_lgb.drop(1,axis=1), submission], axis=1)
display(submission)
```

```python
dt = datetime.datetime.now().strftime('%Y-%m-%d-%H-%M-%S')

submission.to_csv('../data/output/sub_' + dt + '_ensamble_lgb-svr.csv', header=False, index=False)
```
